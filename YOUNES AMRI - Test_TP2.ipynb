2- Write a program to find the number of words in both documents 'a.txt' and 'b.txt' using function.

```python
def NumberOfWords(file):
    _sum=0
    f=open(file,'r')
    for x in f :
        _sum+=len(x.strip("\n,.").split(' '))
    return _sum
        
print('the Number of word in both documents is : ',NumberOfWords('a.txt')+NumberOfWords('b.txt'))
```

3- Write a program to find the number of characters in both documents 'a.txt' and 'b.txt' using function including and execulding spaces

```python
import re
def nbrChar(file,expr):
    s=0
    f=open(file,'r')
    for x in f :
        s+=sum(len(y) for y in re.split(expr,x))
    return s
print('number of characters including spaces : ',nbrChar('a.txt',''),nbrChar('b.txt',''))
print('number of characters execluding spaces : ',nbrChar('a.txt','[^ ]*'),nbrChar('b.txt','[^ ]*')) 
```

6- Write a function and call it to find the Number of stopwords in document 'a.txt' . 

```python
from nltk.corpus import stopwords
def nbrStopwords(file):
    counter=0
    f=open(file,'r')
    for ligne in f :
        counter+=len(list(word for word in re.split(r'[-\s.,;!?"*]+',ligne) if word in stopwords.words('english')))
    return counter
print('the Number of stopwords in document "a.txt" is :',nbrStopwords('a.txt'))
```

7- Write a function and call it to find the Number of special characters '@_+*/=-_-#' in document 'b.txt'.  

```python
from nltk.tokenize import RegexpTokenizer
def NbrOfElement(file,Regexp):
    f=open(file,'r')
    tokenizer = RegexpTokenizer(Regexp)
    return sum(list(len(tokenizer.tokenize(ligne)) for ligne in f))
print('the Number of special characters in document "b.txt" is :',NbrOfElement('b.txt','[\@\+\*\/\=\-\#\?\!\$\%\^\&\(\)\[\]\"\'\.\;,]'))
```

9-Write a function and call it to find all words that occur at least three times in the document 'b.txt'.

```python
import nltk 
from nltk.tokenize import word_tokenize

def Atleast3_Occurence(file):
    f=open(file,'r')
    wordlist=[word for word in word_tokenize(f.read().lower()) if word.isalpha()]
    return dict([(word,counter)for word,counter in nltk.FreqDist(wordlist).items() if counter>=3])

print('All words that occur at least three times in the document "b.txt" is :')
worldlistDict=Atleast3_Occurence('a.txt')
for key in worldlistDict.keys():
         print(key,worldlistDict[key],sep='  ==>  ')     
```

10-Write a function word_freq() that takes a word and the name of text file document as arguments, and computes the frequency of the word in this document. call it using word = 'books'  it document 'b.txt' 

```python
def word_frec(file,word):
    counter=0
    f=open(file,'r')
    tokenizer = RegexpTokenizer(r'\w+')
    wordlist=tokenizer.tokenize(f.read().lower())
    counter=sum([1 for w in wordlist if w==word ])
    return counter
print(word_frec('b.txt','books'))
```

11- Write a program that finds the 20 most frequently occurring words of a text that are not stopwords after Removing Punctuation and special characters and converting all words in document 'b.txt' to miniscule.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

f=open('a.txt','r')
lemmatizer = WordNetLemmatizer()
wordlist=[lemmatizer.lemmatize(word) for word in word_tokenize(f.read().lower()) if word not in stopwords.words('english') and word.isalpha()]
for word,counter in nltk.FreqDist(wordlist).most_common(20):
    print(word,counter,sep='  ==> ')
```
